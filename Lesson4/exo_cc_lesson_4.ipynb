```python
import requests
import re
import json
import pandas as pd
import numpy as np
```

```python
url = "https://open-medicaments.fr/api/v1/medicaments?query=paracetamol&limit=100"

json = requests.get(url).json()
df = pd.DataFrame(json)
```

```python
df.head(5)
```

```python
reg = r"([\D]*)(\d+)(.*),([\w\s]*)"

df_new = df['denomination'].str.extract(reg)
df_new.head(5)
```

```python
df_new2 = df_new.rename(index=str, columns={0: "nom",
                                    1: "dosage",
                                    2: "unite",
                                    3: "type"})
df_new2['mille']=np.where(df_new2['unite'].str.strip()=="g",1000, 1)

df_new2.head(5)
```

```python
df_new2['dosage_new'] = df_new2['dosage'].fillna(0).astype(int)*df_new2['mille'].fillna(0).astype(int)
df_new2.head(5)
```
